```python
import pandas as pd
import pickle
import math
import numpy as np
import re
```

```python
bitcoin = pd.read_csv("../Data/second_comment_bitcoin.csv",index_col = 0, low_memory = False) #low_memory = False
#crypto = pd.read_csv("../Excel/top_comment_crypto.csv",index_col = 0)
```

```python
with open('../list_nameAndTicker.pkl', 'rb') as f: # rb : binary 파일 읽기
    tickers = pickle.load(f)
```

```python
with open('list_2words.pkl', 'rb') as f: # rb : binary 파일 읽기
    two_words = pickle.load(f)
```

```python
new_tickers = []

for i in tickers:
    ticker = i.lower()
    new_tickers.append(ticker)
```

```python
new_tickers
```

```python
def isin_ticker(sentence):
    list_ = []
    for ticker in new_tickers:
        list_.append(any(word.lower() == ticker for word in re.sub(r"[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]","",sentence).split()))
    return any(list_)
```

```python
new_two_words = []

for i in two_words:
    two_word = i.lower()
    new_two_words.append(two_word)
```

```python
def two2one(sentence):
    for two_word in new_two_words:
        if two_word in sentence:
            sentence = sentence.replace(two_word,two_word.split()[0]+"_"+two_word.split()[1])
    return sentence
```

```python
def cleanse(sentence):
    if isinstance(sentence, float) :
        return sentence
    elif ("https://" in sentence) or ('[gif]' in sentence) or ('![gif]' in sentence):
        return math.nan
    else:
        if (sentence == '[deleted]') or (sentence == '[removed]') or (sentence == ''):
            return math.nan
        elif (len(sentence.split())<= 20) and (bool(re.findall('.+\?',sentence)) == False) and (isin_ticker(sentence) == False):
            return sentence
        else:
            return math.nan
```

```python
bitcoin = bitcoin.applymap(cleanse)
#crypto = crypto.applymap(cleanse)
```

```python
bitcoin.to_excel("../Data/second_comment_bitcoin_cl.xlsx",encoding ='utf8')
bitcoin.to_csv("../Data/second_comment_bitcoin_cl.csv")

#crypto.to_excel("../Excel/top_comment_crypto_cl.xlsx",encoding ='utf8')
#crypto.to_csv("../Excel/top_comment_crypto_cl.csv")
```
