```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

from urllib.request import urlopen
import urllib
from tqdm import tqdm

import os
```

```python
headers={'User-Agent': 'Mozilla/5.0'}
list_nameAndTicker = []

for i in range(0,376,25):
    url = 'https://finance.yahoo.com/cryptocurrencies?count=25&offset=' + str(i)
    req = urllib.request.Request(url, headers=headers)
    html = urlopen(req)
    soup = BeautifulSoup(html, "html.parser")
    
    ticker = soup.find_all('a','Fw(600) C($linkColor)')
    name = soup.find_all('td', 'Va(m) Ta(start) Px(10px) Fz(s)')
    
    for j in ticker:
        list_nameAndTicker.append(j.string)
    for k in name:
        list_nameAndTicker.append(k.string)
```

```python
temp = []

for i in list_nameAndTicker:
    a = i.replace("-USD","")
    a = a.replace(" USD","")
    temp.append(a)
```

```python
list_nameAndTicker = temp
list_nameAndTicker
```

```python
list_nameAndTicker.remove('BTC')
list_nameAndTicker.remove('Bitcoin')
```

```python
import pickle

with open('list_nameAndTicker.pkl', 'wb') as f:
    pickle.dump(list_nameAndTicker, f)
```
